```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Churn-Dataset.csv')
```

```python
df
```

```python
df['International plan'] = [0 if x == 'No' else 1 for x in df['International plan']]
df['Voice mail plan'] = [ 0 if x == 'No' else 1 for x in df['Voice mail plan']]
```

```python
df
```

```python
df['Churn'] = [ 0 if x ==False else 1 for x in df['Churn']]
df['Churn']
```

```python
X = df.iloc[:,3:19].values
```

```python
X
```

```python
y = df.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBClassifier
```

```python
x_train,x_test,y_train,y_test =train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
classifier = XGBClassifier()
parameters = {
        'n_estimators': [100, 500],
        'subsample': [0.8, 1.0], 
        'gamma' : [0,1,5],
        'max_depth': [3, 4, 5], 
        'learning_rate': [0.1, 0.3]}
xgboost_cv = GridSearchCV(classifier, parameters, cv = 3, n_jobs = -1, verbose = 2)
```

```python
best = xgboost_cv.best_params_
best
```

```python
xgboost_cv.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred = xgboost_cv.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```
