```python
def onehot_labels(values):
    label_encoder = LabelEncoder()
    integer_encoded = label_encoder.fit_transform(values)
    onehot_encoder = OneHotEncoder(sparse = False)
    integer_encoded = integer_encoded.reshape(len(integer_encoded),1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    return onehot_encoded
```

```python
experience_level = df['experience_level']
experience_level = onehot_labels(experience_level)
experience_level
```

```python
# 3. sıra MI , 4. sıra SE , 1. sıra EN , 2 ise EX
df= df.drop(['experience_level'] , axis = 1)
experience_level_df = pd.DataFrame(data = experience_level , columns=['EN','EX','MI','SE'])
experience_level_df
```

```python
df_concat = pd.concat([experience_level_df , df] , axis = 1 )
df_concat=df_concat.drop(['Unnamed: 0'] , axis = 1)
df_concat
```

```python
salary_currency = df['salary_currency']
salary_columns = list(set(df['salary_currency']))
salary_currency = onehot_labels(salary_currency)
salary_df = pd.DataFrame(data=salary_currency , columns=salary_columns)
salary_df
```

```python
Y = df_concat['salary'].values
```

```python
X = df_concat.drop(['salary'],axis = 1)
X['work_year'] = X['work_year'].astype(np.float32)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
x_train = np.asarray(x_train)
x_test = np.asarray(x_test)
y_train = np.asarray(y_train).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)
```

```python
import tensorflow as tf
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Input(shape=(125,)))
model.add(Dense(512))
model.add(LeakyReLU())
model.add(Dense(1024))
model.add(LeakyReLU())
model.add(Dropout(0.25))
model.add(Dense(1024))
model.add(LeakyReLU())
model.add(Dense(512))
model.add(LeakyReLU())
model.add(Dropout(0.25))
model.add(Dense(1,activation='linear'))
model.compile(optimizer='Nadam',loss=tf.keras.losses.MeanSquaredLogarithmicError(),metrics=['MeanSquaredLogarithmicError'])
```

```python
history = model.fit(x_train , y_train ,epochs=50 , validation_data =(x_test,y_test))
```
